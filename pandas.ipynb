# PANDAS IN PYTHON

```python
import pandas as pd
```

```python
pd.__version__
```

```python
list_s = [1,2,-6,8.9,'data values']
print(list_s)
```

```python
series1 = pd.Series(list_s)
print(series1)
```

```python
type(series1)
```

```python
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
empty_s = pd.Series([])
print(empty_s)
```

assigning index

```python
series3 = pd.Series([1,2,3,4], index = ['a','b','c','d'])
print(series3)
```

declaring a datatype

```python
series4 = pd.Series([5,6,7,8], index =['e','f','g','h'], dtype = float)
print(series4)
```

giving a column name

```python
series5 = pd.Series([73,74,76],index = ['suchi','sukanya','barsha'], name= 'roll numbers')
print(series5)
```

for declaring single (scalar) values

```python
scalar_s = pd.Series(0.5)
print(scalar_s)
```

for declaring same value for different index

```python
scalar_s = pd.Series('mca', index =['suchi','sukanya','barsha'])
print(scalar_s)
```

create from dictionary

```python
dict_s = pd.Series({'a':1, 'b':2})
print(dict_s)
```
